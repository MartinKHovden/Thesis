# Activation functions for neural networks

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-5, 5, 1091)
```

## Sigmoid

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x)*(1 - sigmoid(x))

def tanh(x):
    return (np.exp(2*x)-1)/(np.exp(2*x)+1)

def tanh_derivative(x):
    return 1 - tanh(x)*tanh(x)

def ReLu(x):
    return x*(x>0)

def ReLu_derivative(x):
    ret = 1. * (x>0)
    ret[x==0.0] = np.nan
    return ret 

def leakyReLu(x):
    return np.where(x>0, x, x*0.01)

def leakyReLu_derivative(x):
    ret = np.where(x>0, 1, 0.1)
    ret[x==0.0] = np.nan
    return ret
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,13))

ax[0,0].plot(x_values, sigmoid(x_values), "b")
ax[0,0].plot(x_values, sigmoid_derivative(x_values), "r")
ax[0,0].set_title("Sigmoid", fontsize=25)
ax[0,0].legend(["f(x)", "f'(x)"], fontsize=25)

ax[0,0].tick_params(labelsize=25)
# ax[0,0].set_xlabel("x", fontsize=25)
ax[0,0].yaxis.offsetText.set_fontsize(25)


ax[1,0].plot(x_values, tanh(x_values), "b")
ax[1,0].plot(x_values, tanh_derivative(x_values), "r")
ax[1,0].set_title("Tanh", fontsize=25)
ax[1,0].legend(["f(x)", "f'(x)"], fontsize=25)

ax[1,0].tick_params(labelsize=25)
ax[1,0].set_xlabel("x", fontsize=25)
ax[1,0].yaxis.offsetText.set_fontsize(25)


ax[1,1].plot(x_values, ReLu(x_values), "b")
ax[1,1].plot(x_values, ReLu_derivative(x_values), "r")
ax[1,1].set_title("ReLu", fontsize=25)
ax[1,1].legend(["f(x)", "f'(x)"], fontsize=25)

ax[1,1].tick_params(labelsize=25)
ax[1,1].set_xlabel("x", fontsize=25)
ax[1,1].yaxis.offsetText.set_fontsize(25)


ax[0,1].plot(x_values, leakyReLu(x_values), "b")
ax[0,1].plot(x_values, leakyReLu_derivative(x_values), "r")
ax[0,1].set_title("Leaky ReLu", fontsize=25)
ax[0,1].legend(["f(x)", "f'(x)"], fontsize=25)

ax[0,1].tick_params(labelsize=25)
# ax[0,1].set_xlabel("x", fontsize=25)
ax[0,1].yaxis.offsetText.set_fontsize(25)


plt.savefig("activationFunctions.png",bbox_inches='tight', pad_inces=0.0)
```
