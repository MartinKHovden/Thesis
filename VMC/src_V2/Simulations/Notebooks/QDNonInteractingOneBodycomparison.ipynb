```python
import numpy as np 

densities = np.loadtxt("../Data/quantumDot/OneBody/onebodytest62Slater.txt")
densities = [sum(densities[i:i+4]) for i in range(0, len(densities), 4)]

normalizationFactor = 2.0
normalizedSlater = []
for i in range(len(densities)):
    normalizedSlater.append(densities[i]/((2*i+1)*np.pi*((10/1000)**2)))

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(np.linspace(0,5,len(normalizedSlater)),normalizedSlater, s = 20)
# plt.plot(np.linspace(0,5,len(normalizedSlater)),[np.exp(-r**2)*1.34e6 for r in np.linspace(0,5,len(normalizedSlater))], '--ro' , markersize=1)
plt.legend([ "Exact","VMC: Slater"])
plt.ylim((0, 3.25e7))

plt.show() 

print(len(densities))
```
