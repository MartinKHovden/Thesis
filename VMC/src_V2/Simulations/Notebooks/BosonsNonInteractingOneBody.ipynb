```python
import numpy as np 

densities = np.loadtxt("../Data/bosons/OneBody/bosonsOnebodyTest102Gaussian.txt")

normalizationFactor = 2.0
normalizedSlater = []
for i in range(len(densities)):
    normalizedSlater.append(densities[i]/((2*i+1)*np.pi*((10/1000)**2)))

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(np.linspace(0,5,len(normalizedSlater)),normalizedSlater, s = 20)
plt.plot(np.linspace(0,5,len(normalizedSlater)),[np.exp(-r**2)*6.7e6 for r in np.linspace(0,5,len(normalizedSlater))], '--ro' , markersize=1)
plt.legend([ "Exact","VMC: Slater"])
plt.show() 

print(len(densities))
```

```python
import numpy as np 

densities = np.loadtxt("../Data/bosons/OneBody/bosonsOnebodyTest102GaussianNN.txt")

normalizationFactor = 2.0
normalizedNN = []
for i in range(len(densities)):
    normalizedNN.append(densities[i]/((2*i+1)*np.pi*((10/1000)**2)))

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(np.linspace(0,5,len(normalizedNN)),normalizedNN, s = 20)
plt.plot(np.linspace(0,5,len(normalizedNN)),[np.exp(-r**2)*6.7e6 for r in np.linspace(0,5,len(normalizedNN))], '--ro' , markersize=1)

plt.legend([ "Exact","VMC: Slater-NN"])
plt.show() 

print(len(densities))
```
