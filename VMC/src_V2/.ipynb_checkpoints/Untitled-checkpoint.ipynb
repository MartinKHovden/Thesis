```python
import numpy as np 

densities = np.loadtxt("onebodytest.txt")
```

```python
normalized = []
for i in range(len(densities)):
    normalized.append(densities[i]/((2*i+1)*np.pi*((10/1000)**2)))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.scatter(range(500),normalized[:500], s = 5)
```
